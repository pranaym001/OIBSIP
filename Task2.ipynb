```python
# Step 1: Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Step 2: Create a sample unemployment dataset
# Generating dates from January 2019 to December 2021
date_range = pd.date_range(start='2019-01-01', end='2021-12-31', freq='M')
# Simulating unemployment rates with some random values
np.random.seed(0)  # For reproducibility
unemployment_rates = np.random.uniform(low=3.0, high=15.0, size=len(date_range))

# Creating a DataFrame
unemployment_data = pd.DataFrame({
    'Date': date_range,
    'Unemployment Rate': unemployment_rates
})

# Step 3: Display the first few rows of the dataset
print(unemployment_data.head())

# Step 4: Check for missing values
print(unemployment_data.isnull().sum())

# Step 5: Handle missing values (if any)
unemployment_data.dropna(inplace=True)

# Step 6: Convert 'Date' column to datetime format (already done in creation)

# Step 7: Summary statistics
print(unemployment_data.describe())

# Step 8: Visualize unemployment rate over time
plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Unemployment Rate', data=unemployment_data)
plt.title('Unemployment Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Unemployment Rate (%)')
plt.xticks(rotation=45)
plt.grid()
plt.show()

# Step 9: Analyze the impact of COVID-19
# Assuming the lockdown started around March 2020
pre_covid = unemployment_data[unemployment_data['Date'] < '2020-03-01']
post_covid = unemployment_data[unemployment_data['Date'] >= '2020-03-01']

# Calculate average unemployment rates before and after COVID-19
avg_pre_covid = pre_covid['Unemployment Rate'].mean()
avg_post_covid = post_covid['Unemployment Rate'].mean()

print(f'Average Unemployment Rate Before COVID-19: {avg_pre_covid:.2f}%')
print(f'Average Unemployment Rate After COVID-19: {avg_post_covid:.2f}%')

# Step 10: Visualize the comparison
labels = ['Before COVID-19', 'After COVID-19']
averages = [avg_pre_covid, avg_post_covid]

plt.figure(figsize=(8, 5))
sns.barplot(x=labels, y=averages)
plt.title('Average Unemployment Rate Before and After COVID-19')
plt.ylabel('Average Unemployment Rate (%)')
plt.show()
```
